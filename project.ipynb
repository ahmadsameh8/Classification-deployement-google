```python
import pandas as pd
import numpy as np


csv_file_path = 'drive/MyDrive/output.csv'


df_temp = pd.read_csv(csv_file_path, nrows=1)
columns = df_temp.columns


```

```python
columns
```

```python
cols = []
for i in columns:
  if i not in ["Unnamed: 0","0"]:
    cols.append(i)
```

```python
dtype_dict = {columns[0]: str, columns[-1]: str}

for col in columns[2:-1]:
    dtype_dict[col] = np.float32


data = pd.read_csv(csv_file_path, usecols=cols, dtype=dtype_dict)
print(data.dtypes)
```

```python
data = data.dropna()
data
```

```python
data['Label'] = pd.factorize(data['Label'])[0]
```

```python
len(data['Label'].unique())
```

```python
data['Label'].value_counts()
```

```python
data = data.drop(["0"], axis=1, errors="ignore") # ID Column
```

```python
X = data.drop("Label", axis=1)
y = data["Label"]
```

```python
# MAke Train_test_spilt before anything
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
model = RandomForestClassifier()
rfe = RFE(estimator=model, n_features_to_select=9503)
rfe.fit(X, y)
```

```python

feature_importances = rfe.estimator_.feature_importances_
```

```python
print(feature_importances)
```

```python
results_df = pd.DataFrame({'Feature': X.columns, 'Score': feature_importances})
```

```python



ranked_features =results_df.sort_values(by='Score', ascending=False)
final_features =  ranked_features.head(300)
with open("finalfeatures.txt", "w") as file:
    file.write(str(final_features["Feature"].to_list()))
```

```python
(final_features)
```

```python
X = X[final_features["Feature"]]
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y,random_state=10)
```

```python
from xgboost import XGBClassifier # move with the import
```

```python
xgb = XGBClassifier(n_estimators=5000,
            objective='multi:softprob',
            learning_rate= 0.04,
            max_depth=250,
            gamma = 0,
            colsample_bytree= 0.0009,
            tree_method= 'hist',
            n_jobs = -1,
            random_state= 1)
```

```python
xgb.fit(X_train, y_train, verbose=True) # use sf varable not X_train, note that you have to make train test spilt one more time, as this is a new features
```

```python

y_pred = xgb.predict(X_train)
predictions = [round(value) for value in y_pred]
accuracy = accuracy_score(y_train, predictions)
print("Train Accuracy: %.2f%%" % (accuracy * 100.0))
y_pred = xgb.predict(X_test)
predictions = [round(value) for value in y_pred]
accuracy = accuracy_score(y_test, predictions)
print("Test Accuracy: %.2f%%" % (accuracy * 100.0))
```
